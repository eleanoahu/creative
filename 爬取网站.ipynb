```python
import requests
url ='https://www.douban.com/doulist/11874/'
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.102 Safari/537.36 Edg/104.0.1293.63"}

response = requests.get(url,headers=headers)
print(response)
```

```python
html = response.text
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html.parser")
movie_list = soup.find("div",class_="article")
print(movie_list)
```

```python
movies = movie_list.find_all("div",class_="doulist-item")
movie_names = []
for movie in movies:
    print(movie)
```

```python
movie_names = []
for movie in movies:
    movie_name = movie.find("div",class_ ="title").get_text()
    print(movie_name)
    movie_names.append(movie_name.strip())
```

```python
for name in movie_names:
    print(name)
```

```python
with open("movies.txt","w", encoding="utf-8") as f:
    for name in movie_names:
        f.write(name + '\n')
```
